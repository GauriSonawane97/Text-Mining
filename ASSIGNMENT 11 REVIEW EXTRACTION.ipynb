```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
link="https://www.amazon.in/Nike-Mens-Vapormax-Running-Shoes/product-reviews/B07ZQ71J9J/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page=requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names=soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
len(names)
```

```python
cust_name=[]
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(0)
```

```python
cust_name
```

```python
title=soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:]=[titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:]=[titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating=soup.find_all('i',class_='review-rating')
```

```python
rating
```

```python
rate=[]
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
len(rate)
```

```python
rate.pop(0)
```

```python
rate.pop(0)
```

```python
len(rate)
```

```python
review=soup.find_all("span",{"data-hook":"review-body"})
```

```python
review
```

```python
review_content=[]
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:]=[reviews.rstrip('\n\n') for reviews in review_content]
review_content
```

```python
review_content[:]=[reviews.lstrip('\n\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```
